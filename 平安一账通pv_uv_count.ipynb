```python
import re
import pandas as pd
import xlwt
```

```python
with open("/home/fanjindong/logs/access.log") as f:
    access_data=f.read()
```

```python
# pingan-one
map_app_code={"chongzhi":"充值",
              "jiayouka":"加油卡",
              "shenghuojiaofei":"生活缴费",
              "weizhang":"违章缴费"}
```

```python
# abchina
map_app_code={
    #"dianying":"电影",
              #"zhuanche":"专车",
              #"waimai":"外卖",
              "menpiao":"门票",
#               "piaowu":"票务",
#               "coffee":"咖啡",
              #"jiudian":"酒店"
}
```

```python
df=pd.DataFrame(columns=["appCode","IP","time_day","time_hour"])
add=0
for app_code in map_app_code:
    pattern=re.compile('[0-9].*/api/basis/v1/main/info.*https://abchina.otosaas.com/{}'.format(app_code))
    resp_date=pattern.finditer(access_data)
    for index,value in enumerate(resp_date):
        try:
            value=value.group()
            user_ip=re.match('[0-9].*? ',value).group(0)[:-1]
            time_date=re.findall('[0-9]{2}/May.*? ',value)[0].split(":")
            df.loc[index+add]={"appCode":map_app_code.get(app_code,app_code),
                               "IP":user_ip,
                               "time_day":str(pd.to_datetime(time_date[0])).split()[0],
                               "time_hour":time_date[1]}
        except Exception as why:
            print(value,why)
    add+=index+1
    print(">>>",app_code,index,add+index,"<<<",end='\r')
```

```python
df.groupby(["appCode","time_day"]).count()
```

```python
df_count=df[df["time_day"]].groupby(["appCode","time_day"]).count()
df_count["UV"]=df.drop_duplicates(subset=["appCode","time_day","IP"]).groupby(["appCode","time_day"]).count()["IP"]
del df_count["time_hour"]
df_count.columns=["PV","IP"]
df_count.to_excel("5月份的农行pv_ip数据.xlsx")
```
