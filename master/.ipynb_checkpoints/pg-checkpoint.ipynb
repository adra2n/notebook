```python
# postgresql 不同环境数据同步 v0.1
def postgresql_data_sync(table_from,env_from,env_come,table_come=None,field_as=None):
    """将某个环境的数据表的数据(或部分)同步至另一个环境的另一张数据表中。
    """
    resp_from,keys_from=pg_select("SELECT {field_as}* FROM {table_from};".format(field_as=field_as if field_as else '',table_from=table_from),env_from,field=True)
    if not resp_from:
        print(table_from,table_come,'null')
        return 
    _,keys_come=pg_select("SELECT * FROM {table_come} LIMIT 1;".format(table_come=table_come),env_come,True)
    keys_merge=[key for key in keys_from if key in keys_come]
    keys=', '.join(keys_merge)
    value=', '.join(["?"]*len(keys_merge))
    if table_come is None:
        table_come=table_from
#     return table_come,keys,value
    pg_insert("INSERT INTO {table_come} ({keys}) VALUES ({value});".format(table_come=table_come,keys=keys,value=value),resp_from,env_come)
    print(table_from,table_come,'ok')
```

```python
# postgresql 不同环境数据同步 v0.2
# 每次单条数据的insert，update! 
import traceback
def postgresql_data_sync(table_from,env_from,env_come,
                         table_come=None,
                         field_as=None,
                         method='insert',
                         filter_where='',
                         map_field_value={},
                         resp_from=None,
                         keys_from=None):
    """将某个环境的数据表的数据(或部分)同步至另一个环境的另一张数据表中。
    Args:
        table_from(str):
        env_from(Object):
        env_come(Object):
        table_come(str):
        field_as(str): "app_id as application_id,enterprise_id as organization_id"
        method(str): "insert" or "update"
        filter_where(str): "app_id='10086'"
        map_field_value(dict): {"field":{value:map_value}}
        resp_from(bool): None 源数据表数据
        keys_from(bool): None 源数据表keys
    """
    if table_come is None:
        table_come=table_from
        
    if resp_from is None and keys_from is None:
        resp_from,keys_from=fetch_data_from_table(table_from,env_from,field_as=field_as,filter_where=filter_where)

    if not resp_from:
        print(table_from,table_come,'respFrom is null')
        return 
    
    keys_come=fetch_keys_come_table(table_come,env_come)
    keys_merge=[key for key in keys_from if key in keys_come]
    resp_from=fetch_data_regular(resp_from,map_field_value)
    
    if method=='insert':
        keys=', '.join(keys_merge)
        value=', '.join(["?"]*len(keys_merge))
    elif method=='update':
        #默认用户创建表的首字段为主键，故[0]获取到主键
        main_id=keys_come[0]
        keys_merge.remove(main_id) if main_id in keys_merge else 0
        sql_set = '=%s,'.join(keys_merge) + '=%s'

    for item in resp_from:
        try:
            if method=='insert':
                pg_insert("INSERT INTO {table_come} ({keys}) VALUES ({value});".format(
                    table_come=table_come,keys=keys,value=value),[item],env_come)
            elif method=='update':
                pg_update("UPDATE {table_come} SET {keys} WHERE {main_id}=%s;".format(
                    table_come=table_come, keys=sql_set, main_id=main_id),[item],env_come)
        except Exception as e:
            traceback.print_exc()
    print(table_from,table_come,'ok')
```

```python
def fetch_data_from_table(table_from,env_from,field_as='',filter_where=''):
    """数据获取"""
    resp_from,keys_from=pg_select(
        "SELECT {field_as} FROM {table_from} {filter_where};".format(
            field_as=field_as+',*' if field_as else '*',
            table_from=table_from,
            filter_where='WHERE {}'.format(filter_where) if filter_where else ''
        ),
        env_from,field=True)
    return resp_from,keys_from

def fetch_data_regular(data,map_field_value):
    """数据规整，映射"""
    if map_field_value:
        for item in data:
            for key,value in item.items():
                if key in map_field_value:
                    item[key]=map_field_value[key].get(item[key],item[key])
    return data
    
def fetch_keys_come_table(table_come,env_come):
    """提取新表keys"""
    _,keys_come=pg_select("SELECT * FROM {table_come} LIMIT 1;".format(table_come=table_come),env_come,True)
    return keys_come
```

```python
# postgresql 不同环境数据同步
def postgresql_data_sync(table,env_from,env_come):
    """
    Args:
        table(list):['','','']
        env_from(Object):p.dpg
        env_come(Object):p.spg
    """
    for database in table:
        with open(database,'w') as f:
            with env_from as conn:
                with conn.cursor() as cur:
                    cur.copy_to(f, database)
        with open(database) as f:
            with env_come as conn:
                with conn.cursor() as cur:
                    cur.copy_from(f, database)
        os.system("rm -f {}".format(database))
        resp_from=pg_select("""SELECT * FROM "{}";""".format(database),env_from)
        resp_come=pg_select("""SELECT * FROM "{}";""".format(database),env_come)
        print(resp_from==resp_come,len(resp_from),len(resp_come))
```

```python
def insert_sql_parse(sql):
    pattern = re.compile("\(.+?\)")
    r = re.findall(pattern, sql)
    r[0] = r[0].replace('(', '')
    r[0] = r[0].replace(')', '')
    r[0] = r[0].replace(' ', '')
    keys = r[0].split(',')
    assert len(keys) == len(r[1].split(','))
    return keys


def pg_insert(sql, data, conn):
    """
    封装PostgreSql语句之 INSERT INTO
    Args:
        sql(str):
        data(list):[{},{}]
    Returns:
        out: list of tuples
    """
    assert isinstance(sql, str)
    assert isinstance(data, list)
    assert ";" in sql
    sql = sql.replace('?', '%s')
    keys = insert_sql_parse(sql)
    conn.rollback()
    with conn.cursor() as curs:
        sql_pg = ""
        for item in data:
            value = [item.get(k) for k in keys]
            sql_pg += curs.mogrify(sql, value).decode()

        curs.execute(sql_pg)
        conn.commit()
    return
```

```python
def update_sql_parse(sql):
    sql_list = sql.split(' ')
    sql_list = [v for v in sql_list if '%s' in v]
    sql_clean = ''.join(sql_list)
    sql_clean = sql_clean.replace(';', '')
    sql_clean = sql_clean.replace('=%s', ',')
    keys = [v for v in sql_clean.split(',') if v]
    assert len(keys) > 1
    return keys
def pg_update(sql, data, conn):
    """
    封装PostgreSql语句之 UPDATE
    Args:
        sql(str): "UPDATE pay SET app_id=%s WHERE id=%s;"
        data(list):[{},{}]
    Returns:
        out: list of tuples
    """
    assert isinstance(sql, str)
    assert isinstance(data, list)
    assert ";" in sql
    sql = sql.replace('?', '%s')
    keys = update_sql_parse(sql)
    conn.rollback()
    with conn.cursor() as curs:
        sql_pg = ""
        for item in data:
            value = [item.get(k) for k in keys]
            sql_pg += curs.mogrify(sql, value).decode()
        curs.execute(sql_pg)
        conn.commit()
    return
```

```python
def pg_select(sql, conn, field=None):
    """
    封装PostgreSql语句之 SELECT
    Args:
        sql(str):
    Returns:
        out: list of dict
    datas=pg_select('SELECT key FROM boluome_settlement')
    """
    assert isinstance(sql, str)
    assert ";" in sql
    conn.rollback()
    with conn.cursor() as curs:
        curs.execute(sql)
        keys = curs.description
        data_all = curs.fetchall()
        conn.commit()
    data = []
    for item in data_all:
        data.append({v[0]: item[k] for k, v in enumerate(keys)})
    return data if not field else (data,[v.name for v in keys])
```
