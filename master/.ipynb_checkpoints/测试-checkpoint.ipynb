```python
import math
```

```python
t=10
n=50
lam = 2.897
p_t_n = (lam*t)**n*math.exp(-lam*t)/math.factorial(n)
print('{:.3}%'.format(p_t_n*100))
```

```python
def fetch_probability(lam,t,n):
    return (lam*t)**n*math.exp(-lam*t)/math.factorial(n)
```

```python
result = {}
for n in range(1,100):
    p_t_n = fetch_probability(lam,t,n)
    result[n] = result.get(n-1,1)-p_t_n
```
