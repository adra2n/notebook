```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,16)          # 数据准备
print('x:',x)
y = np.array([1,3,5,7,9,10,12,13,15,16,18,19,20,21,23])/100
print('y:',y)

cof = np.polyfit(x,y,5)           # 调用函数，用 3 次多项式拟合
print(cof)                       # 返回多项式的系数
p=np.poly1d(cof)                  

plt.plot(x,y,'o',x,p(x),lw=2)
```
