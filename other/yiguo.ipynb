```python
def headers_body_fetch(timestamp=None, method=None, request_biz_data=None):
    """
    封装函数之获取接口请求request.post(url,headers=headers,data=body)的headers,body参数
    """
    sign = hashlib.md5('customeCode:{},bizSource:{},timestamp:{},method:{},requestBizData:{}'.format(
        YIGUO_CUSTOME_CODE, YIGUO_BIZ_SOURCE, timestamp, method, request_biz_data).encode('utf-8')).hexdigest()
    body = {'timestamp': timestamp,
            'customeCode': YIGUO_CUSTOME_CODE,
            'method': method,
            'bizSource': YIGUO_BIZ_SOURCE,
            'sign': sign,
            'requestBizData': str(request_biz_data)}
    headers = {'content-type': 'application/x-www-form-urlencoded;charset=UTF-8'}
    return headers, body
```

```python
def dilivery_city_query():
    method='delivery.city.query'
    request_biz_data={}
    timestamp=time.strftime('%Y-%m-%d %H:%M:%S', time.localtime())

    headers, body = headers_body_fetch(timestamp=timestamp, method=method, request_biz_data=request_biz_data)
    resp_data = requests.post(url=YIGUO_HOST, data=body, headers=headers).json()
    return get_in(resp_data,'responseBizData.provList')
# dilivery_city_query()
```

```python
def category_list_query(area_id):
    method='category.list.query'
    request_biz_data={'areaId': area_id}
    sign = hashlib.md5('customeCode:{},bizSource:{},timestamp:{},method:{},requestBizData:{}'.format(
                        custome_code,biz_source,timestamp,method,request_biz_data).encode('utf-8')).hexdigest()
    payload = {'timestamp':timestamp, 
               'customeCode':custome_code,
               'method':method,
              'bizSource':biz_source,
              'sign':sign,
            'requestBizData':str(request_biz_data)}
    headers={'content-type':'application/x-www-form-urlencoded;charset=UTF-8'}
    resp_data=requests.post(url=YIGUO_HOST,data=payload,headers=headers).json()
    return resp_data['responseBizData']
```

```python
def category_list_update_to_mongo(area_id,name):
    category_list=category_list_query(area_id)
    if category_list['categoryList']:
        c.static.shengxian_yiguo_category.update_one({'areaId':area_id,'name':name},
                                                     {'$set':category_list},upsert=True)
```

```python
count=0
for city in dilivery_city_query():
    try:
        count+=1
        print(f"{count},{city['name']}",end='\r')
        category_list_update_to_mongo(city['areaId'],city['name'])
        for city_list in city.get('cityList',[]):
            count+=1
            print(f"{count},{city['name']},{city_list['name']}",end='\r')
            category_list_update_to_mongo(city_list['areaId'],city_list['name'])
            for area_list in city_list.get('areaList',[]):
                count+=1
                print(f"{count},{city['name']},{city_list['name']},{area_list['name']}",end='\r')
                category_list_update_to_mongo(area_list['areaId'],area_list['name'])
    except :
        pass
#     break
```
