```python
import pdb
def consumer():
    r = ''
    while True:
        pdb.set_trace()
        n = yield r
        if not n:
            return
        print('[CONSUMER] Consuming %s...' % n)
        pdb.set_trace()
        r = '200 OK'

def produce(c):
    c.send(None)
    pdb.set_trace()
    n = 0
    while n < 5:
        n = n + 1
        print('[PRODUCER] Producing %s...' % n)
        r = c.send(n)
        print('[PRODUCER] Consumer return: %s' % r)
    c.close()

c = consumer()
pdb.set_trace()
produce(c)
```
