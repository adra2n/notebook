## 驴妈妈price 字段(全量scenic更新)
+ 根据scenic数据表，拿到所有景区ID
+ 根据景区ID，从goods数据表中获取price信息
+ 移除goods数据表中prices过期票价信息

```python
def binary_search(array, t):
    """二分法删除时间t之前的票价列表，返回t时间对应array的索引
    """
    low = 0
    height = len(array) - 1
    while low <= height:
        mid = math.floor((low + height) / 2)

        if len(array[mid]['date']) != 10:
            try:
                array[mid]['date'] = time.strptime(
                    array[mid]['date'],
                    '%Y-%m-%d'
                )
                t = time.strptime(t, '%Y-%m-%d')
            except Exception as e:
                continue

        if array[mid]['date'] < t:
            low = mid + 1
        elif array[mid]['date'] > t:
            height = mid - 1
        else:
            return mid

    return low

def scenic_price(scenic_id):
    today_date=time.strftime('%Y-%m-%d', time.localtime(time.time()))
    price=-1
    market_price = -1
    status=False
    goods=m.loc['menpiao_goods'].find(
        {'placeId':str(scenic_id)},
        {'prices':1,'id':1,'_id':0})
    for good in goods:
        prices=good.get('prices')
        good_id=good.get('id')
        if not prices:
            continue
        #price为票价列表（date，marketprice，sellprice，...）
        price_index = binary_search(prices, today_date)
        
        if price_index < len(prices):
            update_prices=prices[price_index:]
            #删除过期的票价
            m.loc['menpiao_goods'].update_one(
                {'id':good_id},
                {'$set':{'prices':update_prices}})
            for up_price in update_prices:
                if price==-1:#初始化price
                    price=up_price['sellPrice']
                    market_price=up_price['marketPrice']
                    status=True
                elif 0<up_price['sellPrice']<price:#拿到最低price
                    price=up_price['sellPrice']
                    market_price=up_price['marketPrice']
                    status=True
            
        else:
            m.loc['menpiao_goods'].update_one(
                {'id': good_id},
                {'$unset': {'prices': ''}})
    m.loc['menpiao_scenic'].update_one(
        {'id': str(scenic_id)},
        {'$set': {
            'price':price,
            'marketPrice':market_price,
            'status':status
            }
        }
    )
    print({'scenicId':scenic_id,'price':price,'marketPrice':market_price,'status':status})
    return 
if __name__=="__main__":
#     scenic_ids=m.loc['menpiao_scenic'].find({},{'id':1})
#     for scenic_id in scenic_ids:
#         scenic_price(scenic_id['id'])
```
